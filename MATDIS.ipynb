```python
def factorial(n):
    result = 1
    while(n>0):
        result *= n
        n -= 1
    return result

f1 = factorial(25)
print('The result of factorial from 25 is',f1)
```

```python
from itertools import combinations

x = ['TI1','T12','TI3','TI4','TI5','TI6','TE1','TE2','TE3','TE4','TE5','TE6','TE7','TE8']
kombinasi = combinations(x, r=4)
kombinasi = list(kombinasi)
print('Hasil dari kombinasi = ',len(list(combinations(x, r=4))))

i = 1
for data in kombinasi:
    print(f'hasil kemungkinan {i} : {data} ')
    i= i+1
```

```python
kombinasi1 = combinations(x[:6], r=2)
kombinasi2 = combinations(x[6:],r=2)
kombinasi1 = list(kombinasi1)
kombinasi2 = list(kombinasi2)
print('Hasil dari kombinasi TI = ',len(list(combinations(x[:6], r=2))))
print('Hasil dari kombinasi TE = ',len(list(combinations(x[6:], r=2))))
print("Hasil dari total TI dan TE = ",len(list(combinations(x[:6], r=2)))*len(list(combinations(x[6:], r=2))))
i = 1
for TI in kombinasi1:
    for TE in kombinasi2:
        print(f'hasil kemungkinan {i} : {TI+TE} ')
        i= i+1
```
